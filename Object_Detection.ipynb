```python
source_url = "https://res.cloudinary.com/di6nk1mov/image/upload/v1692735921/E-learnig/rahk6f5awmqqyjt90yaj.jpg"
output_img = !yolo predict model=yolov8n.pt source='{source_url}'
# Extract lines starting with "image"
image_lines = [line for line in output_img if line.startswith('image')]

# Regular Expression
import re

# Define a regular expression pattern to capture the text between dimensions and processing time
pattern = r'(\d+x\d+)\s(.*?)\s(\d+\.\d+ms)'

# Extract and print the parts for each line
for line in image_lines:
    match = re.search(pattern, line)
    if match:
        dimensions = match.group(1)  # e.g., "640x640"
        detected_objects = match.group(2)  # e.g., "2 bottles, 1 potted plant, 1 vase"
        processing_time = match.group(3)  # e.g., "528.7ms"
        print("Detected Objects:", detected_objects)
```

```python
# To Translate
from googletrans import Translator

# Create a Translator object
translator = Translator()

# Translate "Detected Objects" to Arabic
translated_text = translator.translate(detected_objects, src='en', dest='ar').text

# Print the translated text
print(translated_text)
```
